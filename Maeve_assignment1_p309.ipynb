### Questions


Recall that the speed of light in vacuum is 
\begin{equation}
c = \frac{1}{\sqrt{\epsilon_0 \mu_0}}, \; \; \; \; \; (1)
\end{equation}
where $\epsilon_0 = 8.85 \times 10^{-12}$ F/m (Farads per meter)$^{*}$ is the electric permittivity of free space and $\mu_0 = 1.26 \times 10^{-6}$ N/A$^2$ (Newtons per Ampere$^2$)$^{**}$ is the magnetic permeability of free space. The electric field created by a point charge $q$ is
\begin{equation}
E = \frac{1}{4 \pi \epsilon_0} \frac{q}{r^2}. \; \; \; \; (2)
\end{equation}
Suppose a charged particle is moving at some fraction of the speed of light in vacuum ($v_q = f c; \; 0 < f < 1$) as seen in reference frame $S$. Reference frame $S^{\prime}$ is co-moving with the point charge, so that $\frac{dx^{\prime}}{dt^{\prime}} = v^{\prime}_q = 0$ in frame $S^{\prime}$. To make the math easier, put $q$ at the origin of $S^{\prime}$.

To understand Einstein's motivation for developing the special theory of relativity, make the <strong>incorrect, erroneous, and wrong</strong> assumption that the Galilean velocity transformation holds true for all values of $f$:
\begin{equation}
\frac{dx^{\prime}}{dt^{\prime}} = \frac{dx}{dt} - v_q \; \; \; (3)
\end{equation}

$^{*}$Farad = Coulomb/Volt: Farad is the unit of capacitance, which measures the amount of charge necessary to increase the electric potential of a system by 1 Volt. Farads/meter measure how much capacitance you get per meter of free space between yourself and a charged particle.

$^{**}$Newtons/Ampere$^2$ = Newtons/(Coulombs/second)$^2$. $\mu_0$ measures how effective inductance is at creating magnetic force in free space.

```python
# Plot here
e_0 = constants.epsilon_0
m_0 = constants.mu_0
q = constants.e
k = 1/(4*np.pi*e_0)


r = np.arange(1,10,.5)
E_field = (k*q)/r**2
    
    
plt.plot(r,E_field)
plt.xlabel("Distance between particles (m)")
plt.ylabel("Electric field (V/m)")
plt.title("Electric field as a function of r")
plt.show()
```

**Explanation:**

dx/dt = light moving in a still frame (S), the speed of light in the moving frame according to a still observer, the speed of light plus the speed of the moving frame according to a still observer.

dx'/dt' = light moving in a moving frame (S')  

Vq = speed if the moving frame by itself also according to the observer.

The speed of light in a moving frame from the perspective of the still frame is smaller than the speed of light in the still frame because the same particle moves in both frames, and in the moving frame, the particle would appear to move slower because the frame it is in is also moving.  To solve for dx'/dt' (the speed of light in a still frame) it follows that you would subtract Vq from dx/dt:

dx/dt = dx'/dt' + Vq

dx'/dt' = dx/dt - Vq

Also, the speed of the moving frame will be some fraction of the speed of light:
Vq = fc where {0<f<1}.  dx/dt is the new value of c, denoted by c'.  We can sub this into the equation:
dx'/dt' =  c'

dx'/dt' = c - Vq

c' = c - cf  ---> c ---> c' = c(1-f)


```python
# Plot here

c = constants.c
f = np.arange(0,1,0.1)
c_prime = c*(1-f)
plt.plot(f,c_prime)
plt.title("c prime as a function of f {0<f<1}")
plt.xlabel("f {0<f<1}")
plt.ylabel("c prime (m/s)")
plt.show()
```
