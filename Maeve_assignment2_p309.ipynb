Recall that in PHYS 207, you were often asked to plot displacement as a function of time. For example, for a baseball thrown directly upward you would plot 
y(t), and for a block on a horizontal spring you would plot 
x(t). Here we will examine how the displacement and time axes "scissor together" at relativistic speeds.

Imagine that frame 
 K' is moving in the 
 +x direction at speed 
v relative to frame K. You are at rest in frame 
K. Let's figure out how the set of (t', x')
 axes would look to you when the frames' origins O 
 and O'
 coincide.

1. Use the fact that x' = 0
 on the t'
 axis to define a line x(t)
 in frame K
 that follows the t'
 axis. Write a function that computes points on the line. v
will be one of your function inputs.

The line will depend on v, how fast K' is moving away from K, and will have a start and stop value.  The time (x-axis) will be a list of values of the start and stop time that our K' moves away at.  Our y axis will be x', which depends on t and v.  Since v = x/t, x = vt, so our y values will depend on the rate at which the frame moves upwards, and the passage of time.  These ideas combined should result in a linear line with a positive slope.  The function should return the x-axis and y-axis list, and when we plot, we need to put in the proper parameters, which will result in two unique arrays for the axes, which will create a unique line tracing the t' axis.

```python
def plot_points(v, start_time, stop_time):
  t_prime = np.arange(start_time, stop_time)
  x_prime = v * t_prime
  return t_prime, x_prime


t_prime_coordinate_list = plot_points(2,0,10)[0]
x_prime_coordinate_list = plot_points(2,0,10)[1]
plt.plot(t_prime_coordinate_list, x_prime_coordinate_list)
```

2. Use the fact that 
 t' = 0 on the x'
 axis to define a line 
 x(t) that traces the x'
 axis in frame K. Write a second function that computes points on this line.

If t'= 0 on the x' axis, then the frames both start at the same origin.  From the still frame K, the axis of x' (vertical axis) should appear to only move upwards with certain velocity.  x' should then be defined as an array that depends on v and has a start and end value.  t' does not change, so we need to make it an array that does not change, (an arrary of zeros).  

```python
def plot_points2(v, start_time, stop_time):
  x_prime = v*np.arange(start_time, stop_time)
  t_prime = 0*np.arange(start_time, stop_time)
  return t_prime, x_prime


t_prime_coordinate_list = plot_points2(2,0,10)[0]
x_prime_coordinate_list = plot_points2(2,0,10)[1]
plt.plot(t_prime_coordinate_list, x_prime_coordinate_list)
```

3. The code below creates a figure with centered (ct,x) axes that have range (-10^10, 10^10). We are using ct
 instead of just t
on our horizontal axis so that both axes have the same units. Use your functions from problems 1 and 2 to overplot the t' and x' axes as a function of ct
 for v= 0.2c. Make your t'
and x' axes red, and label them in red.

```python
# Set up an array of independent variables t that spans -100 to 100 seconds
t = np.linspace(0, 200, num=201, endpoint=True) - 100

# Make a figure
fig, ax = plt.subplots(figsize=(7,7))

# YOUR CODE HERE: Use ax.plot() to plot the t' and x' axes below
c = constants.c
def plot_points(v, start_time, stop_time):
  t_prime = c*np.arange(start_time, stop_time)
  x_prime = 0.2*t_prime
  return t_prime, x_prime

t_prime_coordinate_list = plot_points(0.2*c,-100,100)[0]
x_prime_coordinate_list = plot_points(0.2*c,-100,100)[1]
ax.plot(t_prime_coordinate_list, x_prime_coordinate_list, color = 'red')


def plot_points2(v, start_time, stop_time):
  x_prime = 0.2*c*np.arange(start_time, stop_time)
  t_prime = 0*np.arange(start_time, stop_time)
  return t_prime, x_prime

t_prime_coordinate_list = plot_points2(0.2*c,-100,100)[0]
x_prime_coordinate_list = plot_points2(0.2*c,-100,100)[1]
ax.plot(t_prime_coordinate_list, x_prime_coordinate_list, color = 'red')

# Move the left and bottom "spines" (borders of the plot) to the center
# This makes a math textbook-like set of (x,t) axes instead of a box around the plot
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

# Take out the top and right spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Make the two axes have the same scale
ax.set_aspect('equal')

# Set the axes ranges
ax.set_xlim([-1e10, 1e10])
ax.set_ylim([-1e10, 1e10])

# YOUR CODE HERE: Use ax.text() to add RED labels to the t' and x' axes below
ax.text(0.75,0.40, 'tc prime', horizontalalignment = 'center', verticalalignment = 'center', transform = ax.transAxes).set_color('red')
ax.text(0.60,1.00, 'x prime', horizontalalignment = 'center', verticalalignment = 'center', transform = ax.transAxes).set_color('red')

```
