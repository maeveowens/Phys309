1. Recall the velocity addition example from class: Frank is on a space station, Mary is in a spacecraft moving away from the space station in the +x direction, and she is firing a proton gun at asteroids that are in the spacecraft's path.
Assume the protons leave the proton gun with speed 0.99c in the +x direction. Plot the proton speed measured by Frank as a function of spacecraft speed relative to the space station.
Now suppose the protons leave the gun in a "cone" shape rather than straight along the +x axis, so that their velocity vectors have y and z components. The angle of the proton cone is 15 degrees, and their launch speed is still 0.99c. Plot Frank's measurement of the y-component of the proton velocity at the edge of the cone, as a function of spacecraft speed.

Frank is in the still frame, he measures the proton speed to be u

Mary is in the moving frame in the moving ship, she measures the proton speed to be u_prime = 0.99c

At first the proton only moves in the x-direction (no other components need to be accounted for).

The Lorentz-velocity transformation for an object moving in the x-direction is u = (u' + v) / (1 + (v/c^2)u')
    

```python
#Proton moving in the x-direction only
c = constants.c
u_prime = 0.99*c

v_values = np.arange(0,u_prime,10000)
u_values_x = []
#u_values_y = []

#Make vertical axis array
def u_x(v_values):
  for v in v_values:
    ux = (u_prime + v) / (1 + (v/c**2) * u_prime)
    #uy = ux * np.sin(np.pi / 12)
    u_values_x.append(ux)
    #u_values_y.append(uy)
  return u_values_x#, u_values_y

u_values_x_2 = u_x(v_values)#[0]
#u_values_y_2 = u_x(v_values)[1]
u_values_y_2 = u_values_x_2 * np.int(np.sin(constants.pi / 12))
print(u_values_x_2)
#print(u_values_y_2)

plt.plot(v_values, u_values_x_2)
plt.ylabel("Proton Speed Measured by Frank")
plt.xlabel("Velocity of Ship")
plt.title("Proton speed in x-direction as a function of ship velocity")
plt.show()  
```

Since the velocity of the proton in the y direction is directly proportional to its x-direction velocity, both graphs should have similar curves, but different scales on their vertical axes.  Each u measured in the y direction is the x component multiplied by sin(15 degrees) where 15 degrees = pi/12 rad.  

```python

#Proton moving in the y-direction
#15 degrees = pi/12 rad

u_values_y = []
#Make vertical axis array
def u_y(v_values):
  for v in v_values:
    uy = ((u_prime + v) / (1 + (v/c**2) * u_prime)) * 2 * np.sin(np.pi/12)
    u_values_y.append(uy)
  return u_values_y

u_values_y_2 = u_y(v_values)

plt.plot(v_values, u_values_y_2)
plt.ylabel("Proton Speed Measured by Frank")
plt.xlabel("Velocity of Ship")
plt.title("Proton speed in y-direction as a function of ship velocity")
plt.show()   
```

From JWST point of view, the flashes would be simultaneous because its world line intersects with both flashes world lines at the same point on our space-time axes.
