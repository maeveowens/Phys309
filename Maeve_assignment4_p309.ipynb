2. Assume you are trying to verify the experimental results that led to Einstein's successful explanation of the photoelectric effect. You have two emitters or "photocathodes"; one is made of aluminum and the other is made of gold. On one set of axes, plot the maximum photoelectron kinetic energy in eV that you could measure as a function of frequency (nu) from both photocathodes. Make sure your plot shows the frequency cutoffs below which no photoelectrons will be generated. Label both axes and put a legend on your plot so a reader with no physics experience will know which curve belongs to which metal. You may find this table of work functions helpful.

```python
frequencies = np.linspace(7.5e14,3e15,100)

def find_KE(frequency_array,work):
  kinetic_energy = h*frequency_array - work
  return kinetic_energy

aluminum_work = 4.08 * 1.602e-19
gold_work = 5.1 * 1.602e-19
aluminum_KE = find_KE(frequencies,aluminum_work)
gold_KE = find_KE(frequencies,gold_work)

#Plot
plt.plot(frequencies,aluminum_KE)
plt.plot(frequencies, gold_KE)
plt.axhline(0.0, color = "green")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Kinetic Energy [J]")
plt.title("Kinetic Energy vs. frequency for aluminum and gold")
plt.legend(['Aluminum', 'Gold', 'Frequency cutoff'],loc = "upper left", frameon = True, )
plt.text(1.5e15,0, "*Note Below*", color = 'purple')
```

*Note* Below the green line, kinetic energies for certain frequencies do not exist because they are negative (KE cannot be < 0).  At each intersection, the x-coordinate (frequency) is the minimum frequency at which photoelectrons can be emitted (depending on metal).  Therefore, incident light with frequencies less than this intersection for each metal will not produce photoelectrons.
